# Radial Calendar Plots

Radial calendar plots offer a unique and visually engaging way to display hourly data over a circular timeline. Inspired by Sara Abouebeidâ€™s adaptation of [Adam Heisserer's Energy Calendar Plot](https://www.adamheisserer.com/#/energy-monitoring-calendars/), this approach transforms traditional data visualization into an elegant format that highlights temporal patterns and trends.

Originally designed for energy data in Grasshopper, the radial calendar plot method has been successfully adapted to work with any hourly dataset. Its design emphasizes clarity and aesthetics, making it easier to interpret complex datasets at a glance.

Below is an example of how to use the `plot_radial_calendar` function to create your own radial calendar plot.

## Usage of `plot_radial_calendar`

The `plot_radial_calendar` function creates a radial plot that visualizes hourly data over a circular timeline. Below is an explanation of its key parameters:

- **data**:  
  The input DataFrame containing your data.

- **size_column**:  
  The name of the column in your DataFrame that defines the size of each plotted point. This allows you to visually scale the points based on a specific metric (e.g., power, intensity).

- **colour_column**:  
  The column name that determines the color of each point. This is useful for adding an extra dimension to your visualization (e.g., temperature, irradiance).

- **inner_radius & outer_radius**:  
  These parameters set the inner and outer boundaries of the plot, effectively controlling the thickness of the calendar ring.

- **min_size & max_size**:  
  Define the minimum and maximum size for the points in the plot, allowing you to scale the visual elements according to the data values.

- **bg_color**:  
  Sets the background color of the plot. This can be customized to match your design or presentation preference.

- **line_color**:  
  Specifies the color of the grid or division lines in the plot.

- **month_label_color**:  
  Color used for the month labels around the circle, ensuring they are legible against the background.

- **show_legend**:  
  A boolean flag that, if set to `True`, displays a legend explaining the scale, color mapping, or any relevant metrics.

- **title & subtitle**:  
  These parameters allow you to add a title and a subtitle to the plot, providing context to the viewer.

- **fig_size**:  
  A tuple defining the size of the figure (width, height). This helps in adjusting the visual dimensions according to your output medium.

- **cmap**:  
  Defines the colormap used for the points. For instance, the 'inferno' colormap can be applied for a vivid visualization.

- **attribution_text & attribution_color**:  
  These allow you to set a text (e.g., data source or credits) and its color for proper acknowledgment.

- **dpi**:  
  Dictates the resolution of the plot in dots per inch, which is important for high-quality outputs.

- **glow**:  
  If enabled (`True`), adds a glow effect to the points, enhancing visual appeal.

- **mask_months**:  
  A list of months (represented by their numerical values) that should be masked out (not shown) in the plot. This is useful if you want to focus on specific parts of the year.

This example demonstrates how to set up and call the `plot_radial_calendar` function to visualize your data in an effective and aesthetically pleasing manner. Adjust the parameters as needed to suit the specifics of your data and visualization requirements.

## Load data

```python
import pandas as pd
data = pd.read_csv('data.csv')
```

# Prepare data for plotting

```python
# # Max and min of P
# print("Max of P:", data['P'].max())
# print("Min of P:", data['P'].min())

# # Replace values below 700 with NaN
# data.loc[data['P'] < 400, 'P'] = None
```

## Import function and plot

```python
from radial_plot import plot_radial_calendar


fig,ax = plot_radial_calendar(
    data=data,                                                                      # DataFrame with the data
    size_column='P',                                                                # Column with the size of the points    
    colour_column='H_sun',                                                          # Column with the colour of the points  
    inner_radius=5,                                                                 # Inner radius of the plot   
    outer_radius=25,                                                                # Outer radius of the plot
    min_size=0.1,                                                                   # Minimum size of the points
    max_size=20,                                                                    # Maximum size of the points
    bg_color="black",                                                               # Background color of the plot
    line_color="white",                                                             # Color of the lines
    month_label_color="white",                                                      # Color of the month labels
    show_legend=True,                                                               # Show the legend
    title=f"Radial Calendar for P and H_sun",                                       # Title of the plot
    subtitle="A simple python script to plot radial calendars from hourly data",    # Subtitle of the plot
    fig_size=(10, 10),                                                              # Size of the figure
    cmap='inferno',                                                                 # Colormap for the points
    attribution_text="Data from PVGIS",                                             # Attribution text
    attribution_color="white",                                                      # Color of the attribution text
    dpi=300,                                                                        # Dots per inch of the figure
    glow=True,                                                                      # Add a glow effect to the points
    mask_months = [1,2,3]                                                           # Months to mask (1=January, 2=February, etc.)
)
```

## Save plot

```python
fig.savefig("radial_calendar.png", dpi=300, bbox_inches='tight')
```
